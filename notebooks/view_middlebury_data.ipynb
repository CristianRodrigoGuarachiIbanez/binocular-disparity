```python
%matplotlib inline
import matplotlib.pylab as plt

from disparity.wrangle_data import load_middlebury_dataset

data_dir = '../data/'
```

```python
samples = load_middlebury_dataset('../data', nb_samples=5, max_size=400)
```

```python
n = len(samples)
fig, axes = plt.subplots(n, 4, figsize=(20,3*n))
for i in range(n):
    img_L, img_R, disp_L, disp_R = samples[i]
    axes[i,0].imshow(img_L)
    axes[i,0].axis('off')
    axes[i,0].set_title('left image')
    axes[i,1].imshow(disp_L)
    axes[i,1].axis('off')
    axes[i,1].set_title('left disparity')
    axes[i,2].imshow(img_R)
    axes[i,2].axis('off')
    axes[i,2].set_title('right image')
    axes[i,3].imshow(disp_R)
    axes[i,3].axis('off')
    axes[i,3].set_title('right disparity')
plt.show()
```
